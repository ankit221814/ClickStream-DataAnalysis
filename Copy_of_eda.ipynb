Question 1  : What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each.
(Hint: Compare their scope, techniques, and applications for each.)

Answer:- 1. Artificial Intelligence (AI)

Scope: The broadest field. It aims to create machines or systems capable of performing tasks that typically require human intelligence, such as visual perception, speech recognition, and decision-making.

Techniques: Includes everything from "if-then" logic (Rule-based systems) to advanced neural networks.

Applications: Self-driving cars, Chess-playing computers, and Virtual Assistants like Siri or Alexa.

2. Machine Learning (ML)

Scope: A subset of AI. It focuses on the idea that we can give machines access to data and let them learn for themselves without being explicitly programmed for every specific task.

Techniques: Statistical algorithms such as Linear Regression, Decision Trees, and Support Vector Machines (SVM).

Applications: Email spam filters, product recommendations on Amazon, and credit scoring.

3. Deep Learning (DL)

Scope: A specialized subset of ML inspired by the structure of the human brain (Artificial Neural Networks). It is designed to handle vast amounts of unstructured data.

Techniques: Multi-layered Neural Networks (CNNs for images, RNNs for text). It requires high computational power (GPUs).

Applications: Facial recognition, real-time language translation, and generative AI (like ChatGPT or image generators).

4. Data Science

Scope: An interdisciplinary field that overlaps with AI, ML, and DL. While AI is about building intelligent systems, Data Science is about extracting insights from data using scientific methods.

Techniques: Data cleaning, Exploratory Data Analysis (EDA), visualization, and applying ML models to solve business problems.

Applications: Predicting customer churn, optimizing supply chains, and healthcare analytics.

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

Answer:- 1. Underfitting

Underfitting occurs when a model is too simple to capture the underlying pattern of the data. It's like trying to understand a complex novel by only reading the page numbers.

Symptoms: High error on both the training data and the test data.

Cause (High Bias): The model makes strong assumptions about the data that aren't true (e.g., using a linear model for data that is clearly curved).

Prevention: * Increase model complexity (e.g., use a non-linear model).

Add more features to the dataset.

Reduce regularization.

2. Overfitting

Overfitting occurs when a model is too complex and starts "memorizing" the noise and outliers in the training data rather than the actual pattern. It's like a student who memorizes the specific answers to a practice test but fails the real exam because they didn't understand the concepts.

Symptoms: Very low error on training data but high error on test data.

Cause (High Variance): The model is overly sensitive to small fluctuations in the training set.

Prevention:

Use more training data.

Simplify the model (reduce features).

Use Regularization techniques.

3. The Bias-Variance Tradeoff

This is the central challenge in ML.

Bias is the error introduced by approximating a real-life problem with a simplified model.

Variance is the error introduced by the model's sensitivity to small fluctuations in the training set.

As you increase model complexity, Bias decreases but Variance increases. The "sweet spot" is the point where the total error (Bias + Variance) is at its lowest.

Question 3:How would you handle missing values in a dataset? Explain at least three
methods with examples.
Hint: Consider deletion, mean/median imputation, and predictive modeling.

Answer:- 1. Deletion (Listwise or Dropping)

This involves removing the missing data from the dataset entirely.

Method: You can drop rows that have any missing values or drop entire columns if they have a very high percentage of missing data (e.g., >60%).

Example: In a survey dataset, if a user didn't answer 90% of the questions, that specific row is likely not useful and can be deleted.

Pros/Cons: It is the simplest method, but it can lead to a significant loss of information if the dataset is small.

2. Mean/Median/Mode Imputation

This involves replacing the missing value with a statistical measure of the rest of the column.

Mean: Use for numerical data that is normally distributed (no outliers).

Median: Use for numerical data that is skewed or has outliers (e.g., Salary data).

Mode: Use for categorical data (e.g., "City" or "Gender").

Example: If the "Age" column in a Titanic dataset has 10 missing values, you calculate the average age of the other passengers and fill those 10 spots with that average.

3. Predictive Modeling (Advanced Imputation)

This treats the missing value as a "target variable" and uses other available features to predict what it should be.

Method: Algorithms like KNN (K-Nearest Neighbors) or Iterative Imputation look at similar data points. If three people with similar "Job Titles" and "Education Levels" earn $50k, the model will impute $50k for a similar person with a missing salary.

Example: Using a passenger's "Class" and "Fare" to predict their missing "Age" value.

Pros/Cons: Much more accurate than simple averages, but computationally expensive for large datasets.

Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

Answer:- An imbalanced dataset occurs when the classes within a target variable are not represented equally. In most cases, there is a "majority class" (the dominant group) and a "minority class" (the rare group).

This is extremely common in real-world scenarios like:

Fraud Detection: Out of 10,000 transactions, only 10 might be fraudulent.

Medical Diagnosis: Identifying a rare disease in a large population.

If you train a model on imbalanced data, it will likely achieve high "accuracy" by simply guessing the majority class every time, while completely failing to identify the rare (and usually more important) cases.

Technique 1: Resampling (SMOTE & Random Sampling)

Resampling involves manually adjusting the dataset to balance the classes before training.

Theoretical: * Random Under-sampling: Deleting examples from the majority class. This is fast but risks losing important information.

Random Over-sampling: Duplicating examples from the minority class. This can lead to overfitting because the model "memorizes" the specific duplicates.

SMOTE (Synthetic Minority Over-sampling Technique): Instead of duplicating rows, SMOTE looks at the feature space of the minority class and creates new, synthetic examples that are mathematically similar to existing ones.


Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.
Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.

Answer:- Distance-Based Algorithms: Algorithms like KNN (K-Nearest Neighbors) and SVM (Support Vector Machines) rely on calculating the Euclidean distance between data points. If one feature (e.g., Annual Income: $50,000) has a much larger scale than another (e.g., Age: 25), the distance calculation will be driven almost entirely by the Income, making the "Age" feature irrelevant.
+1

Gradient Descent: Algorithms like Linear Regression, Logistic Regression, and Neural Networks use Gradient Descent to find the minimum of the loss function. If features are on different scales, the "contour" of the loss function becomes an elongated ellipse. This causes the gradient to oscillate and take much longer to converge. Scaling makes the contours more circular, allowing the model to find the optimum faster.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.

Answer:-  1. Label Encoding

Label Encoding assigns a unique integer to each category in alphabetical order or as they appear.

How it works: If you have a column "Size" with values Small, Medium, and Large, they become 0, 1, and 2.

The "Rank" Trap: The main issue is that algorithms might interpret these numbers as having a mathematical order (2>1>0). This is fine if the data is Ordinal (where order matters), but misleading if it is not.

Example: Education Level (High School < Bachelor's < Master's < PhD).

2. One-Hot Encoding

One-Hot Encoding creates a new binary column (0 or 1) for every unique category in the original column.

How it works: If you have a column "Color" with Red, Blue, and Green, it creates three new columns. If a row is "Red," it gets a 1 in the Red column and a 0 in the others.

The "Dimensionality" Problem: If a column has 100 unique categories (e.g., Zip Codes), One-Hot Encoding will add 100 new columns to your dataset, which can make it slow and memory-intensive.

Example: Gender, Country, or Marital Status.

```python
#Question 7: Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons? Dataset: https://github.com/MasteriNeuron/datasets.git Include your Python code and output in the code box below.)

#answer:-

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the dataset
# Replace the URL with the local path if downloaded
df = pd.read_csv('https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv')

# 2. Data Cleaning
# Remove the row with the '1.9' rating (data entry error)
df = df[df['Rating'] <= 5]
# Drop rows with missing ratings for accurate calculation
df.dropna(subset=['Rating'], inplace=True)

# 3. Group by Category and calculate mean Rating
cat_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# 4. Visualization
plt.figure(figsize=(12, 8))
sns.barplot(x=cat_rating.values, y=cat_rating.index, palette='viridis')
plt.title('Average App Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.xlim(3.5, 5) # Zoom in to see differences clearly
plt.show()

# 5. Output results
print("Top 3 Highest Rated Categories:\n", cat_rating.head(3))
print("\nBottom 3 Lowest Rated Categories:\n", cat_rating.tail(3))

```

```python
#Question 8: Titanic Dataset a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened? b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival? Dataset: https://github.com/MasteriNeuron/datasets.git (Include your Python code and output in the code box below.)

#Answer:-

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"
df = pd.read_csv(url)

# --- Part (a): Survival by Passenger Class (Pclass) ---
pclass_survival = df.groupby('Pclass')['Survived'].mean() * 100

# --- Part (b): Survival by Age Group ---
# 1. Handle missing values in Age by using the median
df['Age'] = df['Age'].fillna(df['Age'].median())

# 2. Create Age Groups
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')
age_survival = df.groupby('AgeGroup')['Survived'].mean() * 100

# --- Visualization ---
plt.figure(figsize=(12, 5))

# Plot for Pclass
plt.subplot(1, 2, 1)
sns.barplot(x=pclass_survival.index, y=pclass_survival.values, palette='Blues_d')
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Percentage (%)')
plt.xlabel('Class (1 = 1st, 2 = 2nd, 3 = 3rd)')

# Plot for Age Group
plt.subplot(1, 2, 2)
sns.barplot(x=age_survival.index, y=age_survival.values, palette='RdYlGn')
plt.title('Survival Rate: Children vs Adults')
plt.ylabel('Survival Percentage (%)')

plt.tight_layout()
plt.show()

# Display Numerical Output
print("Survival Rate by Class (%):")
print(pclass_survival)
print("\nSurvival Rate by Age Group (%):")
print(age_survival)
```

```python
#Question 9: Flight Price Prediction Dataset
#a) How do flight prices vary with the days left until departure? Identify any exponential price
#surges and recommend the best booking window.
#b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
#consistently cheaper/premium, and why?
#Dataset: https://github.com/MasteriNeuron/datasets.git
#(Include your Python code and output in the code box below.)

#Answer:- import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
# Link: https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
try:
    df = pd.read_csv(url)
except:
    # If the URL is unavailable, ensure the file 'flight_price.csv' is in your directory
    df = pd.read_csv('flight_price.csv')

# --- Part (a): Price variation with days left ---
# We calculate the average price for each day remaining until departure
days_price = df.groupby('days_left')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=days_price, x='days_left', y='price', marker='o', color='blue')
plt.title('Average Flight Price vs Days Left Until Departure')
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price (INR)')
plt.gca().invert_xaxis()  # Inverting x-axis to show days approaching departure
plt.grid(True)
plt.show()

# --- Part (b): Airline Comparison for Delhi-Mumbai Route ---
# Filtering the data for the specific route: Delhi to Mumbai
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

plt.figure(figsize=(10, 6))
sns.boxplot(data=route_df, x='airline', y='price', palette='Set2')
plt.title('Flight Price Comparison Across Airlines (Delhi to Mumbai)')
plt.xlabel('Airline')
plt.ylabel('Price (INR)')
plt.show()

# Calculate median prices for numerical comparison
print("Median Prices for Delhi-Mumbai Route:")
print(route_df.groupby('airline')['price'].median().sort_values())
```

```python
#Question 10: HR Analytics Dataset
#a). What factors most strongly correlate with employee attrition? Use visualizations to show key
#drivers (e.g., satisfaction, overtime, salary).
#b). Are employees with more projects more likely to leave?
#Dataset: https://github.com/MasteriNeuron/datasets/blob/e0da5fc18ebf4e3dc90950b932d6bded61e0ee49/hr_analytics.csv


#Answer:-
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv"
df = pd.read_csv(url)

# --- Part (a): Key Drivers of Attrition ---
# We calculate correlation for numeric features against 'left' (1=Left, 0=Stayed)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlations = numeric_df.corr()['left'].sort_values(ascending=False)

# Visualization 1: Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='RdBu', fmt='.2f')
plt.title('Correlation of HR Factors with Attrition')
plt.show()

# Visualization 2: Satisfaction vs Attrition
plt.figure(figsize=(8, 5))
sns.boxplot(x='left', y='satisfaction_level', data=df, palette='Set1')
plt.title('Impact of Satisfaction Level on Attrition')
plt.xticks([0, 1], ['Stayed', 'Left'])
plt.show()

# --- Part (b): Project Count Analysis ---
plt.figure(figsize=(8, 5))
sns.countplot(x='number_project', hue='left', data=df, palette='viridis')
plt.title('Employee Attrition by Number of Projects')
plt.xlabel('Number of Projects Assigned')
plt.ylabel('Count of Employees')
plt.show()

print("Correlation of features with Attrition:\n", correlations)
```
